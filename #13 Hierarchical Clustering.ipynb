```python
import matplotlib.pyplot as plt
import numpy as np
A= np.array([[5,3], [10,12], [24,10], [30,30], [85,70], [71,80], [60,78], [70,55]])
```

```python
plt.scatter(A[:,0],A[:,1])
plt.show
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
x = linkage(A, 'centroid')
dendrogram(x)
plt.show()
```

```python
x = linkage(A, 'single')
dendrogram(x)
plt.show()
```

```python
x = linkage(A, 'average')
dendrogram(x)
plt.show()
```

```python
x = linkage(A, 'complete')
dendrogram(x)
plt.show()
```
